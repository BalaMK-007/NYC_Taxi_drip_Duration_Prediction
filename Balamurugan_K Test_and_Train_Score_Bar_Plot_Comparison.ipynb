# Plotting Bar plots with obtained score values of models.

```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

All this values are taken from the models created in other files

```python
linear_train_score = 1.0
```

```python
linear_test_score = 1.0
```

```python
knn_train_score = 0.7589457835569253
```

```python
knn_test_score = 0.9976648945419085
```

```python
#Train score values in x,y
x=["linear_train_score", "knn_train_score"]
y=[1.0, 0.7589457835569253]
```

```python
#Plotting Bar Plot
plt.figure(dpi=100)
plt.bar(x,y)
plt.xlabel("Models")
plt.ylabel("Train Scores")
plt.show()
```

```python
#Test score values in x,y
x=["linear_test_score", "knn_test_score"]
y=[1.0, 0.9976648945419085]
```

```python
plt.figure(dpi=100)
plt.bar(x,y)
plt.xlabel("Models")
plt.ylabel("Test Scores")
plt.show()
```

Linear Train score is slightly higher then the knn Train Score. Where Linear test score is almost equal to the knn test score.
